```python
# read csv
df = pd.read_csv('data/kaggle_survey_2017_2021.csv',low_memory=False, header=None)
df.head()
```

```python
# obtain all questions
question = (df.iloc[0] + '_' + df.iloc[1])
print(question)
```

```python
# concat first 2 rows as new column header
df.columns = (df.iloc[0] + '_' + df.iloc[1])
df.head()
```

```python
df.isnull().sum() #Calculating null values
```

```python
# checking data in year and changing dtype to int
df['-_Year'] = df['-_Year'].astype('int64')
df.loc[1:, '-_Year'].unique()
```

```python
# visualise responses by year
year_count = df.loc[1:, '-_Year'].value_counts()
year_count = year_count.sort_index()  
year_count.plot(kind='bar', rot = 0, color = 'blue')
plt.xlabel('Year')
plt.ylabel('Responses')
plt.title('Number of Responses by Year')
plt.xticks(rotation=0)
plt.show()
```

```python
#Handling Columns
df['Q1_What is your age (# years)?'] = df['Q1_What is your age (# years)?'].fillna('40-44') #Imputing missing values in 'Age' with the median
print('Age is null:', df.iloc[1:, 2].isnull().sum())
```

```python
# check data in duration and change dtype to int
df['Time from Start to Finish (seconds)_Duration (in seconds)'] = df['Time from Start to Finish (seconds)_Duration (in seconds)'].fillna(0)
df['Time from Start to Finish (seconds)_Duration (in seconds)'] = df['Time from Start to Finish (seconds)_Duration (in seconds)'].astype('int64')
```
